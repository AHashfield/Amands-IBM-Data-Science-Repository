```python
# loop to find the duplicates
i=0
while i in range(len(rawfile)):
    counter=i
    loopcounter=counter
    #print("starti: ", i, "start counter: ", counter, "start loopcounter: ",loopcounter)
    #print(rawfile.iat[counter,5])
    countetal = 1
    nameforetal = rawfile.iat[counter,5]
    while (loopcounter+1 < len(rawfile)) and (rawfile.iat[(loopcounter+1),3]== rawfile.iat[(counter),3]):
        # print("intial #: ", rawfile.iat[counter,3], "next #: ",rawfile.iat[loopcounter+1,3])
        if countetal<3:
            rawfile.iat[counter,5] = rawfile.iat[counter,5]+" / "+rawfile.iat[loopcounter+1,5] # for up to 3 duplciates, concat the Party NAmes
        else:
            rawfile.iat[counter,5] = nameforetal +" / et al" # if greter than 3 dups, then concat "/ et al"
        rawfile.iat[loopcounter+1,5]="remove" # for teh duplicate rows, chagne "contract Party" to "rmove"
        # print(rawfile.iat[counter,5])
        loopcounter=loopcounter+1
        countetal=countetal+1
        # print("loop counter: ",loopcounter, "countetal: ", countetal)
    if i==loopcounter: # increment i - eitehr to the next row or to the end row where loop counter is
        i=i+1 
    else:
        i=loopcounter
    #print("end i: ",i)
print("final i", i)
print("final counter ",counter)

            
        
```
